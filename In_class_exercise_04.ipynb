```python
# Write your code here

from six.moves import urllib
import pandas as pd

txt_array = []
link = "https://raw.githubusercontent.com/unt-iialab/info5731_spring2021/main/class_exercises/01-05-1%20%20Adams%20v%20Tanner.txt"
text_file = urllib.request.urlopen(link)

for line in text_file:
  decode_data = line.decode("utf-8").replace('\r\n', '')
  if decode_data: txt_array.append(decode_data)
df = pd.DataFrame (txt_array, columns = ['Sample Text'])
print(df)

```

**NUMBER** **OF** **SENTENCES**

```python
df['Number of Sentences'] = df['Sample Text'].apply(lambda x: len(str(x).split(".")))
df
```

**NUMBER** **OF** **WORDS** 


```python
df['Number of Words'] = df['Sample Text'].apply(lambda x: len(str(x).split(" ")))
df
```

**NUMBER** **OF** **CHARACTERS**

```python
df['Number of Characters'] = df['Sample Text'].str.len()
df
```

**AVERAGE** **WORD** **LENGTH**

```python
def avg_word(sentence):
  words = sentence.split()
  if len(words)!=0:
    return (sum(len(word) for word in words)/len(words))
  else:
    return None

df['Average Word Length'] = df['Sample Text'].apply(lambda x: avg_word(x))
df
```

**NUMBER** **OF** **STOP** **WORDS**

```python
from nltk.corpus import stopwords

import nltk
nltk.download('stopwords')
stop = stopwords.words('english')

df['Number Of Stop Words'] = df['Sample Text'].apply(lambda x: len([x for x in x.split() if x in stop]))
df
```

**NUMBER** **OF** **SPECIAL** **CHARACTERS**


```python
def special_characters(sentence):
  special = 0
  for value in range(len(sentence)):
    if not((sentence[value].isalpha()) and (sentence[value].isdigit())):
        special = special + 1
  return special

df['Number of special characters'] = df['Sample Text'].apply(lambda x: special_characters(x))
df
```

**NUMBER** **OF** **NUMERICS**

```python
df['Number Of Numerics'] = df['Sample Text'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
df
```

**NUMBER** **OF** **UPPERCASE** **WORDS**

```python
df['Number of UpperCase Words'] = df['Sample Text'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
df
```

**1.2** **Basic** **Text** **Pre**-**processing** **of** **text** **data**

Lower Casing


```python
df['Number Of LowerCase Words'] = df['Sample Text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df
```

**Punctuational** **Removal**

```python
df['Removed Punctuational Text'] = df['Sample Text'].str.replace('[^\w\s]','')
df
```

**Stop** **Words** **Removal**

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
df['Stop Words Removed Text'] = df['Sample Text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df
```

**Frequent** **Words** **Removal**


```python
freq_words = pd.Series(' '.join(df['Stop Words Removed Text']).split()).value_counts()[:10]
print(freq_words)
freq_words_list = list(freq_words.index)
df['Frequent Words Removed Text'] = df['Stop Words Removed Text'].apply(lambda x: " ".join(x for x in x.split() if x not in freq_words_list))
df
```

Rare Words Removal


```python
rare_words = pd.Series(' '.join(df['Stop Words Removed Text']).split()).value_counts()[-10:]
print(rare_words)
rare_words_list = list(rare_words.index)
df['Rare Words removed Text'] = df['Frequent Words Removed Text'].apply(lambda x: " ".join(x for x in x.split() if x not in rare_words_list))
df
```

**Spelling** **Correction**

```python
from textblob import TextBlob
df['Spelling Correction'] = df['Rare Words removed Text'].apply(lambda x: str(TextBlob(x).correct()))
df
```

**Tokenization**

```python
import nltk
nltk.download('punkt')
df['Tokenization']= df['Spelling Correction'].apply(lambda x: TextBlob(x).words)
df
```

**Stemming**

```python
from nltk.stem import PorterStemmer
st = PorterStemmer()
df['After Stemming'] = df['Tokenization'].apply(lambda x: " ".join([st.stem(word) for word in x]))
df
```

**Lemmatization**

```python
from textblob import Word
import nltk
nltk.download('wordnet')

df['After Lemmatization'] = df['After Stemming'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
df
```

**1.3 Save all the clean sentences to a csv file (one column, each raw is a sentence) after finishing all the steps above.**

```python
df.to_csv('cleaned_data.csv',index=False)
```

**1.4 Advance Text Processing**

Calculating Term frequencies


```python
tf1 = (df['After Lemmatization']).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
tf1.columns = ['words','tf']
tf1
```

**Top-10 1 gram**

```python
import itertools
import collections
from nltk import ngrams
from nltk.tokenize import word_tokenize

words_sentence = []
for sentence in df['After Lemmatization']:
  words_sentence.append(word_tokenize(sentence))
aftercleaning = [x for x in words_sentence if x != []]
iterations = list(itertools.chain.from_iterable(aftercleaning))
one_grams = ngrams(iterations, 1)
collections.Counter(one_grams).most_common(10)

```

Top 10- 2 grams

```python
bi_grams = nltk.bigrams(iterations)
frequency_dist = nltk.FreqDist(bi_grams)
(frequency_dist.most_common(10))
```

Top 10 - 3 grams


```python
tri_grams = nltk.trigrams(iterations)
frequency_dist = nltk.FreqDist(tri_grams)
frequency_dist.most_common(10)

```
